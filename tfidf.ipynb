# TF-IDF

Article classification problem:
* a news article is given to you
* classify it into one ot the companies (Tesla, Apple, etc.)

Bag of Words model:
* we'll take an article,  design its vocabulary

Vocabulary:
* all the words
* from all the news articles

We will take count of indicidual word in a given article.

just by looaking at this vector:
* article 1 and article 2 are Apple articles.
* because, they iclude iphone and itunes terms.

article 3 and article 4:
* include musk term, 15 and 3
* this one is for Tesla 

other term, market, price, investor:
* these are very generic terms 
* that can appear in any article.

The presence of these terms in your bag of words model:
* can suppress the meaningful words.

For Apple:
* iphone, itunes is the relevant terms

For Tesla:
* musk and gigafactory are relevant terms.

These other generic terms:
* suppresssing the influence of these relevant terms 

Article 1 and Article 3, "that, price, market, investor" terms:
* are present with equal amount.

If you give these vectors to computer:
* computer thinks, article 1 and article 3 is similar.
* but, article one is from Apple
* the third one is from Tesla.

other terms are kind of equal, there are 4 ot them:
* there could be 100 such terms in yoru vector.
* your computer can think, these article are similar
* but it is not case.

How do you ignore these terms:
* you don't comletely ignore them

That is a stop word:
* remove all the stop words.

How about all these terms: price market, investor:
* stop words can help
* but, certain extent.

We figure out:
* how many documents the term that appeared.
* that in 3 documents out of 4 

Not about the individual word count in a given document. 
* how many documents this term appeared.

that appeared in 3 doc, gigafactory is 1, iphone is 2

This is something:
* called document frequency

Document Frequency (DF):
* Number of times term t is represent in all docs.

if this number is higher:
* if some term is appearing in majority of documents. 
* we should lower its influence
* because, it's like a generic term.

is some word is appearing in only few documents:
* such as Gigafactory, only 1 document
* then that terms is important 
* we need to give it a high score.

We need to come up with come scoring mechanism here.

The scoring mechanisms should be such that:
* you know higher the term appears in all documents
* the influence would be lower.
* You can do that
* by doing inverse document frequency.

That terms is 3, so you invert it.
* the number goes down.

You don't simple inversion but you do this formula:
* total number of documents / number of documents term t is present in

For example:
* that-> 4/3 = 1.33

Denominator:
* how many times that appeared in all the documents are 3.
* total number of documents are 4.

* gigafactory -> 4/1 = 4
* iphone ->4/2 = 2

The gigafactory term appears in less number of documents:
* the score you are getting is higher -> 4

This is called IDF inverse document frequency.

There is something additioanl to this formula:
* which is called you take a log of this.
* log ( total number of documents / number of documents term t is present in )

if the term appears in more documents:
* the final score for that will be lower
* because you want to reduce the influence.

You will come up with this kind of vector:
* which will be called IDF vector

We don't use IDF:
* we use TF - IDF

Take into account the word frequency

We don't want to just take a word frequency:
* like bag of words
* because somer articles could be bigger.

* there is 5000 words in article
* there is 10 words in small article

You want to be fair and kind of normalize things:
* total number of term t is present in document 
* which is just a simple word count that we took in bag of words model
* divided by total number of tokens in that document
* that we will call term frequency

here t is the term and d is that document:
* document in which you are counting that term frequency.

For example:
* the term frequncy in article 1 for the term market 
* well be 48 divided by total number of words in that article.
* total number of words in that article are 1000.

the term frequency for market in article 1 :
* will be 48 divided by 1000 

TF-IDF is just a multiplication of these two terms.

that, price -> will have lower number.

Machine learning models understand only numbers:
* you can use a map function.

You can create a new columns applying transformation on existing column using Pandas.

```python
df['label_num'] = df.label.map({
    'Household': 0,
    'Books':1,
    'Electronics':2,
    'Clothing & Accessories': 3
})
df.head()
```

split train and test 

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df.text, # it is your dependent variable
    df.label_num, #your independent variable, like a target variable
    test_size = 0.2, #20% samples will go to test dataset
    random_state = 2022, #random state parameter randomly initializing it to num
    stratify = df.label_num  # if you have inbalanced data 
)
```

Use few classifiers from sklearn to train my model:
* KNN
* Random Forest
* Naive Bayes

What is the performance of these classifiers?

Library for classifiers

sklearn.pipeline is to create a pipeline: 
* where in a first stage -> you can have your TF-IDF vector
* the second stage -> you have your KNN classifier

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
```

```python
clf = Pipeline([
    ('vectorizer_tfidf', TfidfVectorizer()),
    ('KNN', KNeighborsClassifier())
])
clf.fit(X_train, y_train)
# do prediction on the x test dataset
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.naive_bayes import MultinomialNB

clf = Pipeline([
    ('vectorizer_tfidf', TfidfVectorizer()),
    ('Multi NB', MultinomialNB())
])
clf.fit(X_train, y_train)
# do prediction on the x test dataset
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```

almost similar

random forest classifier

```python
from sklearn.ensemble import RandomForestClassifier

clf = Pipeline([
    ('vectorizer_tfidf', TfidfVectorizer()),
    ('Random Forest', RandomForestClassifier())
])
clf.fit(X_train, y_train)
# do prediction on the x test dataset
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```
